```python
# Odds of Opportunity

import random

events = ('Enemy', 'Trap', 'NPC', 'Rand', 'Clue', 'Loot', 'SD')

for x in range(0, 7):
    event = events[x]
    
    odds = random.choice (['0,1,1,2,3,5,7,10','1,1,2,3,5,9,10,13','1,2,3,4,7,10,11,15',
                         '4,5,9,10,13,16,16,18','5,7,10,11,15,16,18,19','9,10,13,15,16,18,19,19',
                         '10,11,15,16,18,19,19,20','11,13,16,16,18,19,19,22','16,16,18,19,19,20,20,26'])
    li = list(odds.split(","))
    opportunity = random.choice(li)
    roll = random.randrange(1,21,1)
    
    if(roll < int(opportunity)):
        exists = "YES"
    else:
        exists = "NO"    
    print(event + ":" + "(" + opportunity + "-{0:.0%}".format(int(opportunity)/20) + ")" + 
          "\t\tRolled: " + str(roll) + "\t" + exists)
    
# Odds of Opportunity
```
